```python
import pandas as pd
import os

df = pd.read_csv("../../data/raw/ISIC_2019_Training_GroundTruth.csv")
print(df.head())

print(df.columns)

# Calculate class distribution (summing one-hot encoded columns)
classes = ['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC', 'UNK']
class_counts = df[classes].sum()
print(class_counts)

def image_exists(img_id):
    return os.path.exists(f"../../data/raw/images/{img_id}.jpg")

df['exists'] = df['image'].apply(image_exists)
print(df['exists'].value_counts())
```
