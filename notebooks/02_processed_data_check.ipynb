# Processed Data Check

This notebook visualizes the processed images to ensure they are ready for training.
- Check dimensions (300x300)
- Check class distribution
- Visualize samples


```python
import os
import cv2
import matplotlib.pyplot as plt
import glob
import numpy as np

PROCESSED_DIR = '../data/processed'
SPLITS = ['train', 'val', 'test']

def count_images(dir_path):
    counts = {}
    total = 0
    if not os.path.exists(dir_path):
        return counts, total
        
    for cls in os.listdir(dir_path):
        cls_path = os.path.join(dir_path, cls)
        if os.path.isdir(cls_path):
            n = len(glob.glob(os.path.join(cls_path, '*.jpg')))
            counts[cls] = n
            total += n
    return counts, total

for split in SPLITS:
    split_dir = os.path.join(PROCESSED_DIR, split)
    counts, total = count_images(split_dir)
    print(f"{split.upper()}: {total} images")
    print(counts)
    print("-"*20)
```

```python
def show_samples(split='train', num_samples=5):
    split_dir = os.path.join(PROCESSED_DIR, split)
    classes = [d for d in os.listdir(split_dir) if os.path.isdir(os.path.join(split_dir, d))]
    
    fig, axes = plt.subplots(len(classes), num_samples, figsize=(num_samples*3, len(classes)*3))
    
    for i, cls in enumerate(classes):
        cls_dir = os.path.join(split_dir, cls)
        images = glob.glob(os.path.join(cls_dir, '*.jpg'))
        
        if len(images) > num_samples:
            images = np.random.choice(images, num_samples, replace=False)
            
        for j, img_path in enumerate(images):
            img = cv2.imread(img_path)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            
            if len(classes) == 1:
                ax = axes[j]
            else:
                ax = axes[i, j]
                
            ax.imshow(img)
            ax.axis('off')
            if j == 0:
                ax.set_title(cls)
                
    plt.tight_layout()
    plt.show()

show_samples('train', 5)
```
